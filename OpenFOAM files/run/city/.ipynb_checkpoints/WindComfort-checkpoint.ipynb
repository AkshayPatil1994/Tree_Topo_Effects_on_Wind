```python
#!/usr/bin/env python
# coding: utf-8

from imageio.core.request import URI_FILENAME
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
from scipy.interpolate import griddata
import pyvista as pv

import seaborn as sns
```

```python
# # COMFORT MISMATCH PLOTS
# 

print('VELOCITY DIFFERENCE') 
print(' ')

U10 = 2.09 #velocity inflow 10 m height [m/s]
kappa =0.41
z0 = 0.2 # roughness length [m]
ustar = U10*kappa/(np.log((4+z0)/z0))     # shouldn not be changed!!
Uph = ustar/kappa*np.log((1.75+z0)/z0)     # shouldn not be changed!!
print ("Uph = ", Uph)
```

```python
# sampled results from postProcessing > surfaces ##broadleaf conifer ##U_pedestrianH U_z6Normal U_z9Normal U_z12Normal U_z15Normal U_z18Normal
surface_U1 = pv.read('./LoD2/postProcessing/surfaces/552/U_pedestrianH.vtk') ###---------need to be changed---------##
U1_mesh1 = pv.PolyData(surface_U1)
windcomfort = U1_mesh1
windcomfort['magnitude'] = np.linalg.norm(windcomfort['U'], axis=1)
windcomfort_scalars = windcomfort.point_data['magnitude']
```

```python
print(windcomfort_scalars.min())
print(windcomfort_scalars.max())
```

```python
# windcomfort_ = windcomfort
scalars = windcomfort_scalars
scalars[windcomfort['magnitude'] < 2.5] = 1 
scalars[windcomfort['magnitude'] >= 2.5] = 2 
```

```python
print(scalars.min())
print(scalars.max())
```

```python
sargs = dict(
    color='k',
    font_family='courier',
    title_font_size=20,
    label_font_size=18,
    fmt="%.1f",
    height=0.05, 
    width=0.6, 
    vertical=False, 
    position_x=0.2, 
    position_y=0.065,
    title='Velocity Magnitude [m/s]',
    # n_colors=21
    )

plotter.add_mesh(windcomfort, scalars=scalars, scalar_bar_args=sargs, cmap=['blue', 'yellow']) #cmap="plasma_r"
plotter.show_bounds(xlabel='x [m]', ylabel='y [m]', color='black')
#plotter.add_text('VELOCITY LoD1', font='courier', font_size=9, position='upper_edge')
plotter.view_xy()   # plot the xy-plane
plotter.enable_zoom_style()   # enable to zoom in the image

# show and save the created subplots:  ###---------need to be changed---------##
plotter.show(screenshot='./LoD3_windComfort.png')
```

```python
sargs = dict(
    color='k',
    font_family='courier',
    title_font_size=20,
    label_font_size=18,
    n_labels=20,
    fmt="%.1f",
    height=0.05, 
    width=0.6, 
    vertical=False, 
    position_x=0.2, 
    position_y=0.065,
    title='Velocity Magnitude [m/s]',
    # n_colors=21
    )

# plotter set-up (create 2 subplots): bad pedestrian vs. good thermal comfort 
plotter = pv.Plotter(shape=(1, 1), window_size=[800, 1200], border=False)
plotter.set_background("w")
pv.set_plot_theme("document")

boring_cmap = plt.cm.get_cmap("RdYlBu_r", 21)
```
